## Create a uniform distribution

```python
uniform_dist = uniform(sample_range[:,0],sample_range[:,1])
```

```python
#initialization of LHS
lh = pyDOE.lhs(76,samples=10000)
```

```python
# Creating samples
samples = uniform_dist.ppf(lh)
samples.shape
```
