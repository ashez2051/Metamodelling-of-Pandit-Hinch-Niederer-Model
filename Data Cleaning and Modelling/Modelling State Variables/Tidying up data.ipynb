```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dict_1 = pd.read_pickle("0-999")
dict_2 = pd.read_pickle("1000-3000")
dict_3 = pd.read_pickle("3000-5000")
dict_4 = pd.read_pickle("5000-6000")
dict_5 = pd.read_pickle("6000-10000")
```

## combining all dictionaries

```python
states = {}
for d in [dict_1, dict_2, dict_3, dict_4, dict_5]: 
    states.update(d)
```

```python
states
```

## Creating a dataframe, removing all simulations that did not yield an AP

```python
final_data=pd.DataFrame()
no_AP = []
AP = []

for i in range(len(states)):
    data = pd.DataFrame(states[i].T)
#     data.columns = ["V(mV)","Na_i(mM)","Ca_i(mM)","Ca_SR(mM)","K_i(mM)","TRPN(mM)","m","h","j","r","s","s_slow","r_ss","s_ss","y","z","Q_1","Q_2","Q_3",
#         "z_1","z_2","z_3"]
    
    if sum(data[0].values >1) >0:
        AP.append(i)
        final_data=final_data.append(data,ignore_index=False)
    else:
        no_AP.append(i)
        
```
